# Functions

```python
reg = re.compile(r'[a-zA-Z]')
upper_case = re.compile(r'((?<=[\.\?!]\s)(\w+)|(^\w+))')

def capital(string):
    return(string.group().capitalize())

def is_english_word(word): 
    if reg.match(word):
        return True
    else:
        return False

def map_stopwords(word,reverse=False):
    for k,v in stopwords.items():
        word=word.replace(k,v) if not reverse else word.replace(v,k)
    return word

def replacer(s, index, newstring, reverse, nofail=False):
    if ((is_english_word(s[index]))and not reverse):
        return s
    if not nofail and index not in range(len(s)):
        raise ValueError(f"index outside given string: {index}")
    newstring=newstring if (index==0 or reverse) else newstring.lower()
    if index < 0: 
        return newstring + s
    if index > len(s):
        return s + newstring
    return s[:index] + newstring + s[index + 1:]

def translate(string,reverse=False):
    idx=0
    string=map_stopwords(normalize(string),reverse) #Mapping common stopwords and normalizing
    string=string.replace("\ufeff","")
    string=string.replace("?",u'\u061F') if (reverse==True) else string.replace(u'\u061F',"?")
    
    if not reverse:
        string=string.replace("انوالو","involve")
        while(idx < len(string)):
            if(idx>0 and string[idx]==u'\u0644' and ((idx+1>=len(string)-1) or string[idx+1]==' ') and (idx-1==0 or string[idx-1]!=' ') and (idx-2<=0 or string[idx-2]==' ')):
                string=string.replace(u'\u0644',"il")
                idx+1
            if string[idx] in dic:
                try:
                    if(idx==0 or string[idx-1]==' '):
                        string=replacer(string, idx, dic.get(string[idx], {}).get('start'),reverse)
                    elif(idx==(len(string)-1) or string[idx+1]==' '):
                        string=replacer(string, idx, dic.get(string[idx], {}).get('end'),reverse)
                    else:
                        string=replacer(string, idx, dic.get(string[idx], {}).get('mid'),reverse)
                except:
                    string=replacer(string, idx, dic.get(string[idx]),reverse)
            idx+=1
        
        string=string.replace('۔','.')
        string=upper_case.sub(capital, string)
        return string
    
    else:
        string=string.lower()
        string=re.sub("^o","ا",string)
        string=re.sub("\so"," ا",string)
        string= string.replace("o","و")
        string=re.sub("^i","ا",string)
        string=re.sub("\si"," ا",string)
        string= string.replace("i",u'\u06CC')
        string=string.replace('gh',u'\u06CC')
        string=string.replace('ai',u'\u06CC')
        
        
        while(idx < len(string)):
            if string[idx] in dic2:
                #middle
                if(idx>0 and idx<(len(string)-1) and string[idx+1]!=' '):
                    if(string[idx]== 'd' and string[idx+1]=='d'):
                        string=replacer(string, idx, u'\u0688',reverse)
                        string=replacer(string,idx+1,'',reverse)
                        idx+=1
                    elif(string[idx]=='i'):
                        string=replacer(string, idx, u'\u0626',reverse)
                    else:
                        string=replacer(string, idx, dic2.get(string[idx]),reverse)
                elif((idx>0 and (string[idx]=='y' or string[idx]=='i')) and string[idx-1]==u'\u0627' ): #ay/iy = 'ے'
                    string=replacer(string,idx,u'\u06D2',reverse)
                    string=replacer(string,idx-1,'',reverse)
                else:
                    string=replacer(string, idx, dic2.get(string[idx]),reverse)    
            idx+=1
        return string
```

# Final Two Driver Methods

```python
def roman2urdu(word):
    return translate(word,True)
    
def urdu2roman(word):
    return translate(word,False)
```
